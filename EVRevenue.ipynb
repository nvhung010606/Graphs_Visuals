```python
import matplotlib.pyplot as plt
import numpy as np

# Set font
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['DejaVu Sans', 'Arial', 'Helvetica']

# Data - EV/Revenue multiples from your comp table
companies = ['CoreWeave', 'Datadog', 'Palantir', 'CrowdStrike', 'MongoDB', 
             'Twilio', 'Okta', 'DocuSign', 'Zscaler', 'HubSpot', 'Snowflake']

ev_revenue = [35.6, 20.4, 103.6, 30.7, 11.5, 23.3, 16.1, 12.9, 35.1, 27.1, 20.7]

# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 6))

# Create bar chart
bars = ax.bar(companies, ev_revenue, color='#D9A527', width=0.7, edgecolor='none')

# Styling
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)

# Remove grid
ax.grid(False)

# Set labels and title
ax.set_ylabel('EV / 2025E Revenue (x)', fontsize=13, fontweight='bold')
ax.set_title('EV / Revenue  Snowflake vs. High-Growth Software Peers', 
             fontsize=16, fontweight='bold', pad=20)

# Set y-axis limits
ax.set_ylim(0, 115)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=11)

# Add value labels on top of bars
for i, (bar, value) in enumerate(zip(bars, ev_revenue)):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 1.5,
            f'{value:.1f}x',
            ha='center', va='bottom', fontsize=10, fontweight='bold')

# Highlight Snowflake bar with different color
snowflake_index = companies.index('Snowflake')
bars[snowflake_index].set_color('#4A90E2')

plt.tight_layout()
plt.show()

# Print summary
print("\n" + "="*80)
print("EV / REVENUE MULTIPLES - SNOWFLAKE vs HIGH-GROWTH SOFTWARE PEERS")
print("="*80)
print(f"\n{'Company':<20} {'EV / 2025E Revenue':<20}")
print("-"*80)
for company, ev in zip(companies, ev_revenue):
    marker = "  SNOWFLAKE" if company == "Snowflake" else ""
    print(f"{company:<20} {ev:.1f}x{marker}")
print("="*80)

# Calculate statistics
avg_ev_revenue = np.mean(ev_revenue)
median_ev_revenue = np.median(ev_revenue)
snowflake_ev = ev_revenue[snowflake_index]

print("\nKEY VALUATION INSIGHTS:")
print("-"*80)
print(f"Snowflake EV/Revenue:        {snowflake_ev:.1f}x")
print(f"Peer Group Average:          {avg_ev_revenue:.1f}x")
print(f"Peer Group Median:           {median_ev_revenue:.1f}x")
print(f"Snowflake vs Average:        {((snowflake_ev/avg_ev_revenue - 1) * 100):+.1f}%")
print(f"Snowflake vs Median:         {((snowflake_ev/median_ev_revenue - 1) * 100):+.1f}%")
print("\nValuation Context:")
print(f" Snowflake trades at {snowflake_ev:.1f}x 2025E revenue")
print(f" Below peer average of {avg_ev_revenue:.1f}x (discount of {abs((snowflake_ev/avg_ev_revenue - 1) * 100):.1f}%)")
print(f" Significantly below Palantir (103.6x), Zscaler (35.1x), and CoreWeave (35.6x)")
print(f" In line with Datadog (20.4x), both consumption-based business models")
print(f" Premium to MongoDB (11.5x) and DocuSign (12.9x)")
print(" Moderate valuation reflects balanced risk-reward for high-growth data platform")
print("="*80)
```
