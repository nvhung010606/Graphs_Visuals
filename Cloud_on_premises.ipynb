```python
import matplotlib.pyplot as plt
import numpy as np

# Years for the chart
years = ['2018', '2020', '2023', '2025E', '2028E']

# Data based on research (percentages)
on_prem = [35, 27, 20, 15, 10]
cloud = [65, 73, 80, 85, 90]

# Create figure with more space
fig, ax = plt.subplots(figsize=(12, 7))

# Width of bars
width = 0.6
x = np.arange(len(years))

# Create stacked bars
p1 = ax.bar(x, on_prem, width, label='On-Prem / Legacy', color='#8B4513', alpha=0.8)
p2 = ax.bar(x, cloud, width, bottom=on_prem, label='Cloud / Modern Platforms', color='#4A90E2', alpha=0.9)

# Styling
ax.set_ylabel('Percentage of Analytics Workloads', fontsize=12, fontweight='bold')
ax.set_title('Analytics Workloads: On-Prem vs Cloud Migration', fontsize=14, fontweight='bold', pad=25)
ax.set_xticks(x)
ax.set_xticklabels(years, fontsize=11)
ax.set_ylim(0, 110)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add percentage labels on bars
for i in range(len(years)):
    # On-prem label
    if on_prem[i] > 5:
        ax.text(i, on_prem[i]/2, f'{on_prem[i]}%', 
                ha='center', va='center', fontweight='bold', fontsize=11, color='white')
    
    # Cloud label
    ax.text(i, on_prem[i] + cloud[i]/2, f'{cloud[i]}%', 
            ha='center', va='center', fontweight='bold', fontsize=11, color='white')

# Add legend
ax.legend(loc='upper left', fontsize=10, frameon=True, fancybox=True, shadow=True)

# Add grid
ax.set_axisbelow(True)
ax.yaxis.grid(True, linestyle='--', alpha=0.3)
ax.set_yticks([0, 20, 40, 60, 80, 100])
ax.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])

# Add Cloud Migration Trend arrow
ax.annotate('', xy=(0.5, 107), xytext=(4.5, 107),
            arrowprops=dict(arrowstyle='->', lw=2.5, color='#4A90E2'))
ax.text(2.5, 109, 'Cloud Migration Trend', ha='center', fontsize=11, 
        style='italic', color='#4A90E2', fontweight='bold')

# Creative Key Insight Box - positioned prominently in the middle-right area
# Using a larger, more eye-catching design
textstr = 'KEY INSIGHT\n\nBy 2028, ~90% of analytics\nworkloads expected to run\non cloud platforms\n\n Massive migration\n   opportunity for SNOW'
props = dict(boxstyle='round,pad=1', facecolor='#FFE5B4', alpha=0.95, 
             edgecolor='#FF8C00', linewidth=2.5)

# Position it between the bars at 40-60% height
ax.text(3.2, 55, textstr, fontsize=11.5, ha='center', va='center',
        bbox=props, fontweight='bold', linespacing=1.5,
        color='#333333')

# Add a second callout for the 2023 data point
callout_text = '2023: 80% Cloud\n(Inflection Point)'
props2 = dict(boxstyle='round,pad=0.6', facecolor='lightgreen', alpha=0.9, 
              edgecolor='darkgreen', linewidth=2)
ax.text(2, 90, callout_text, fontsize=10, ha='center', va='center',
        bbox=props2, fontweight='bold', color='darkgreen')

# Add annotation arrow pointing to 2023 bar
ax.annotate('', xy=(2, 85), xytext=(2, 88),
            arrowprops=dict(arrowstyle='->', lw=2, color='darkgreen'))

plt.tight_layout()
plt.show()

# Print data table
print("\n" + "="*70)
print("CLOUD VS ON-PREMISES ANALYTICS WORKLOAD DISTRIBUTION")
print("="*70)
print(f"{'Year':<15} {'On-Prem/Legacy':<25} {'Cloud/Modern':<25}")
print("-"*70)
for i, year in enumerate(years):
    print(f"{year:<15} {on_prem[i]}%{' '*22} {cloud[i]}%")
print("="*70)
print("\nKey Market Insights:")
print(" Rapid migration from on-premises to cloud-based analytics platforms")
print(" 2023 marked inflection point: 80% of workloads now cloud-based")
print(" Legacy systems still represent substantial migration runway (~20% in 2023)")
print(" By 2028, on-premises expected to account for only ~10% of workloads")
print("\nDrivers of Cloud Migration:")
print(" Cost efficiency: 15-50% savings vs on-premises infrastructure")
print(" Scalability and elasticity for dynamic workloads")
print(" Reduced maintenance burden")
print(" Access to advanced AI/ML capabilities")
print(" Faster deployment and innovation cycles")
print("\nSnowflake's Opportunity:")
print(" Remaining on-premises workloads = large addressable market")
print(" Cloud-native architecture positions SNOW to capture migration wave")
print(" Multi-cloud strategy aligns with enterprise hybrid approaches")
```
