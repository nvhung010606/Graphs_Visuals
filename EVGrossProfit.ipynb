```python
import matplotlib.pyplot as plt
import numpy as np

# Set font
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['DejaVu Sans', 'Arial', 'Helvetica']

# Data - EV/Gross Profit multiples from your comp table
companies = ['CoreWeave', 'Datadog', 'Palantir', 'CrowdStrike', 'MongoDB', 
             'Twilio', 'Okta', 'DocuSign', 'Zscaler', 'HubSpot', 'Snowflake']

ev_gross_profit = [1.0, 20.0, 124.9, 50.0, 17.0, 8.0, 7.0, 6.0, 21.0, 9.0, 27.0]

# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 6))

# Create bar chart
bars = ax.bar(companies, ev_gross_profit, color='#D9A527', width=0.7, edgecolor='none')

# Styling
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)

# Remove grid
ax.grid(False)

# Set labels and title
ax.set_ylabel('EV / Gross Profit (x)', fontsize=13, fontweight='bold')
ax.set_title('EV / Gross Profit  Snowflake vs. High-Growth Software Peers', 
             fontsize=16, fontweight='bold', pad=20)

# Set y-axis limits - adjusted to show the full range including Palantir
ax.set_ylim(0, 135)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=11)

# Add value labels on top of bars
for i, (bar, value) in enumerate(zip(bars, ev_gross_profit)):
    height = bar.get_height()
    # Special positioning for Palantir (the outlier)
    if value > 100:
        ax.text(bar.get_x() + bar.get_width()/2., height + 3,
                f'{value:.1f}x',
                ha='center', va='bottom', fontsize=10, fontweight='bold')
    else:
        ax.text(bar.get_x() + bar.get_width()/2., height + 1.5,
                f'{value:.1f}x',
                ha='center', va='bottom', fontsize=10, fontweight='bold')

# Highlight Snowflake bar with different color
snowflake_index = companies.index('Snowflake')
bars[snowflake_index].set_color('#4A90E2')

plt.tight_layout()
plt.show()

# Print summary
print("\n" + "="*80)
print("EV / GROSS PROFIT MULTIPLES - SNOWFLAKE vs HIGH-GROWTH SOFTWARE PEERS")
print("="*80)
print(f"\n{'Company':<20} {'EV / Gross Profit':<20}")
print("-"*80)
for company, ev in zip(companies, ev_gross_profit):
    marker = "  SNOWFLAKE" if company == "Snowflake" else ""
    print(f"{company:<20} {ev:.1f}x{marker}")
print("="*80)

# Calculate statistics (excluding Palantir outlier for meaningful average)
ev_gp_without_outlier = [x for i, x in enumerate(ev_gross_profit) if companies[i] != 'Palantir']
avg_ev_gp = np.mean(ev_gp_without_outlier)
median_ev_gp = np.median(ev_gross_profit)
snowflake_ev_gp = ev_gross_profit[snowflake_index]

print("\nKEY VALUATION INSIGHTS:")
print("-"*80)
print(f"Snowflake EV/Gross Profit:        {snowflake_ev_gp:.1f}x")
print(f"Peer Average (ex-Palantir):       {avg_ev_gp:.1f}x")
print(f"Peer Group Median:                {median_ev_gp:.1f}x")
print(f"Snowflake vs Average:             {((snowflake_ev_gp/avg_ev_gp - 1) * 100):+.1f}%")
print(f"Snowflake vs Median:              {((snowflake_ev_gp/median_ev_gp - 1) * 100):+.1f}%")
print("\nValuation Context:")
print(f" Snowflake trades at {snowflake_ev_gp:.1f}x gross profit")
print(f" Above peer average of {avg_ev_gp:.1f}x (excluding Palantir outlier) by {((snowflake_ev_gp/avg_ev_gp - 1) * 100):.1f}%")
print(f" In line with Datadog (20.0x) and Zscaler (21.0x)")
print(f" Below CrowdStrike (50.0x) but above most other peers")
print(f" Palantir (124.9x) is extreme outlier - AI hype and government contracts")
print(" Moderate-to-premium valuation on gross profit basis")
print("="*80)
```
