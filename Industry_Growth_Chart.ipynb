```python
import matplotlib.pyplot as plt
import numpy as np

# Actual market data from research
years = [2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030, 2031, 2032]

# Market size in USD billions (actual and projected)
market_size_billions = [
    19.04,   # 2020 (actual)
    24.5,    # 2021 (actual - average of sources)
    29.0,    # 2022 (estimated based on CAGR)
    33.05,   # 2023 (actual)
    37.8,    # 2024 (actual - average of sources)
    45.0,    # 2025 (projected)
    54.5,    # 2026 (projected)
    66.5,    # 2027 (projected)
    81.8,    # 2028 (projected - conservative estimate)
    100.0,   # 2029 (projected)
    118.5,   # 2030 (projected)
    132.0,   # 2031 (projected)
    147.19   # 2032 (projected)
]

# Convert to index with 2020 = 100
index_2020_base = [(value / market_size_billions[0]) * 100 for value in market_size_billions]

# Create the chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the line
ax.plot(years, index_2020_base, linewidth=2.5, color='#2E86AB', marker='o', markersize=6)

# Fill area under the curve
ax.fill_between(years, index_2020_base, alpha=0.15, color='#2E86AB')

# Add a vertical line to separate actual from projected data
ax.axvline(x=2024.5, color='gray', linestyle='--', linewidth=1, alpha=0.5)
ax.text(2024.5, max(index_2020_base) * 0.95, 'Projected ', 
        ha='left', va='top', fontsize=9, style='italic', color='gray')

# Styling
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.5)

# Labels and title
ax.set_title("Industry Revenue Growth  Cloud Data & Analytics", 
             fontsize=14, fontweight='bold', pad=20)

# Add annotations for key data points
ax.annotate('Base Year\n2020 = 100', 
            xy=(2020, index_2020_base[0]), 
            xytext=(2020.5, 150),
            fontsize=9,
            bbox=dict(boxstyle='round,pad=0.5', facecolor='lightblue', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3', lw=1.5))

ax.annotate(f'2028E\nIndex: {index_2020_base[8]:.0f}', 
            xy=(2028, index_2020_base[8]), 
            xytext=(2027, index_2020_base[8] + 100),
            fontsize=9,
            bbox=dict(boxstyle='round,pad=0.5', facecolor='lightgreen', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3', lw=1.5))

ax.annotate(f'2032E\nIndex: {index_2020_base[-1]:.0f}', 
            xy=(2032, index_2020_base[-1]), 
            xytext=(2030.5, index_2020_base[-1] + 150),
            fontsize=9,
            bbox=dict(boxstyle='round,pad=0.5', facecolor='#FFD700', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3', lw=1.5))

# Set x-axis to show all years
ax.set_xticks(years)
ax.set_xticklabels(years, rotation=45)

# Add CAGR annotation
cagr_text = "CAGR 2020-2032: ~18-22%"
ax.text(0.02, 0.98, cagr_text, transform=ax.transAxes,
        fontsize=10, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8, edgecolor='gray'))

plt.tight_layout()
plt.show()

# Print the data table for reference
print("\n" + "="*70)
print("CLOUD DATA & ANALYTICS MARKET - INDUSTRY REVENUE GROWTH")
print("="*70)
print(f"{'Year':<10} {'Market Size (USD B)':<20} {'Index (2020=100)':<20}")
print("-"*70)
for i, year in enumerate(years):
    status = "(E)" if year > 2024 else ""
    print(f"{year}{status:<6} ${market_size_billions[i]:<18.2f} {index_2020_base[i]:<20.1f}")
print("="*70)
print("\nKey Insights:")
print(f" Market growing from ${market_size_billions[0]:.2f}B in 2020 to ${market_size_billions[-1]:.2f}B by 2032")
print(f" Represents a {index_2020_base[-1]:.0f}% increase over 12 years (Index from 100 to {index_2020_base[-1]:.0f})")
print(f" Compound Annual Growth Rate (CAGR): ~20% across the period")
print(f" 2028 projected market size: ${market_size_billions[8]:.1f}B (Index: {index_2020_base[8]:.0f})")
print("\nMarket Drivers:")
print(" Rising adoption of AI and machine learning analytics")
print(" Increasing data volumes and need for scalable solutions")
print(" Shift from on-premises to cloud-based infrastructure")
print(" Real-time analytics demand across industries")
print(" Digital transformation initiatives post-COVID-19")
```
