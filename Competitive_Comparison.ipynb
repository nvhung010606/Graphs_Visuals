```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Set the font
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['DejaVu Sans', 'Arial', 'Helvetica']

# Create figure and axis with more width for legend
fig, ax = plt.subplots(figsize=(15, 7))
ax.axis('tight')
ax.axis('off')

# Define the data
features = [
    'Multi-Cloud Agnostic Deployment',
    'Decoupled Storage & Compute',
    'Data Sharing Ecosystem',
    'AI/ML Native Integration',
    'Transactional + Analytical (UPSERT)',
    'Marketplace / Third-Party Apps',
    'Consumption-Based Usage Pricing'
]

companies = ['Snowflake', 'Databricks', 'MongoDB', 'Palantir', 'Datadog']

# Data matrix ( = full support,  = no support, Partial/Limited = partial support)
data = [
    ['', '', '', '', ''],
    ['', 'Partial', '', '', ''],
    ['', '', '', '', ''],
    ['', '', 'Limited', '', ''],
    ['', 'Partial', '', '', ''],
    ['', '', '', '', ''],
    ['', '', '', '', '']
]

# Combine features and data
table_data = []
for i, feature in enumerate(features):
    row = [feature] + data[i]
    table_data.append(row)

# Column headers
col_labels = ['Feature / Capability'] + companies

# Create the table
table = ax.table(cellText=table_data, colLabels=col_labels,
                cellLoc='center', loc='center',
                colWidths=[0.28, 0.13, 0.13, 0.13, 0.13, 0.13])

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 2.8)

# Color scheme
header_color = '#1A5490'  # Dark blue
snowflake_color = '#D6EAF8'  # Light blue
other_color = '#FFFFFF'  # White
feature_color = '#E8E8E8'  # Light gray
border_color = '#95A5A6'  # Gray

# Style header row
for i, col in enumerate(col_labels):
    cell = table[(0, i)]
    cell.set_facecolor(header_color)
    cell.set_text_props(weight='bold', color='white', fontsize=12)
    cell.set_edgecolor(border_color)
    cell.set_linewidth(2)

# Style data cells
for i in range(len(table_data)):
    for j in range(len(col_labels)):
        cell = table[(i+1, j)]
        cell.set_edgecolor(border_color)
        cell.set_linewidth(1.5)
        
        # First column (features) - bold and left-aligned
        if j == 0:
            cell.set_text_props(weight='bold', ha='left', fontsize=11)
            cell.set_facecolor(feature_color)
        # Snowflake column - highlight
        elif j == 1:
            cell.set_facecolor(snowflake_color)
            cell.set_text_props(fontsize=14, weight='bold')
            # Color the checkmarks
            if table_data[i][j] == '':
                cell.set_text_props(color='#27AE60', fontsize=20, weight='bold')
        # Other company columns
        else:
            cell.set_facecolor(other_color)
            cell.set_text_props(fontsize=13)
            # Color coding
            if table_data[i][j] == '':
                cell.set_text_props(color='#27AE60', fontsize=18, weight='bold')
            elif table_data[i][j] == '':
                cell.set_text_props(color='#E74C3C', fontsize=18, weight='bold')
            elif table_data[i][j] in ['Partial', 'Limited']:
                cell.set_text_props(color='#E67E22', fontsize=11, style='italic', weight='bold')

# Add title
title = fig.text(0.5, 0.96, 'Competitive Feature Comparison: Cloud Data Platforms', 
                ha='center', fontsize=16, fontweight='bold', 
                color='#1A5490', family='serif')

# Add legend - positioned BELOW the table to avoid covering content
legend_elements = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#27AE60', 
               markersize=10, label=' = Full Support'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#E74C3C', 
               markersize=10, label=' = Not Available'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='#E67E22', 
               markersize=10, label='Partial/Limited = Partial Support')
]
legend = ax.legend(handles=legend_elements, loc='lower center', 
                  bbox_to_anchor=(0.5, -0.15), fontsize=10, frameon=True,
                  fancybox=True, shadow=True, title='Legend', title_fontsize=11,
                  ncol=3)  # Display in 3 columns horizontally

# Add footer note
fig.text(0.5, 0.01, 'Analysis based on publicly available product documentation and feature sets as of 2024-2025', 
         ha='center', fontsize=9, style='italic', color='#7F8C8D', family='serif')

plt.tight_layout()
plt.subplots_adjust(top=0.92, bottom=0.08)
plt.show()

# Print summary
print("\n" + "="*100)
print("COMPETITIVE FEATURE COMPARISON TABLE - SNOWFLAKE vs KEY COMPETITORS")
print("="*100)
print(f"\n{'Feature / Capability':<40} {'Snowflake':<12} {'Databricks':<12} {'MongoDB':<12} {'Palantir':<12} {'Datadog':<12}")
print("-"*100)
for i, feature in enumerate(features):
    row = f"{feature:<40} {data[i][0]:<12} {data[i][1]:<12} {data[i][2]:<12} {data[i][3]:<12} {data[i][4]:<12}"
    print(row)
print("="*100)

print("\n" + "="*100)
print("KEY COMPETITIVE INSIGHTS:")
print("="*100)
print("\nSnowflake's Differentiation (7/7 features):")
print("   Multi-Cloud Agnostic Deployment - UNIQUE advantage")
print("   Fully Decoupled Storage & Compute - Superior architecture")
print("   Data Sharing Ecosystem - Proprietary capability")
print("   AI/ML Native Integration - Full support")
print("   Transactional + Analytical - Hybrid workload capability")
print("   Marketplace / Third-Party Apps - UNIQUE ecosystem")
print("   Consumption-Based Pricing - Flexible cost model")

print("\nCompetitor Gaps:")
print("   Databricks: 2.5/7 (Limited marketplace, partial storage/compute)")
print("   MongoDB: 2/7 (Strong in transactional, limited AI/ML)")
print("   Palantir: 1/7 (AI/ML only)")
print("   Datadog: 2/7 (AI/ML and pricing model)")

print("\nSnowflake's Moat:")
print("   Only platform with true multi-cloud portability")
print("   Unique data sharing and marketplace ecosystem")
print("   Superior architecture for analytics workloads")
print("="*100)
```
